```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname='TW-Kai-98_1.ttf')

df = pd.read_csv('Fire generating electric power gas.csv',encoding='utf-8')

#選取[燃油]資料

fuel = df[df['發電方式']=='燃油']

#將資料撈出[年度]、[排放量(萬噸)]
data = fuel.loc[:, ['年度','排放量(萬噸)']]

#將[年度]設定成X軸
data = data.set_index('年度')

#產生圖表
fig = data.plot(kind='bar').get_figure()
plt.title('火力發電-燃油 溫室氣體排放量')
```
